```python
#1 to 10 print
def show(i):
  if i <=10:
    print(i)
    show(i+1)
show(1)
```

```python
#reverse 1 to 10 print
def show(i):
  if i <=10:
    show(i+1)
    print(i)

show(1)
```

```python
#array print
def printarr(a,i): #i=0,1,2,3,4
  if i<len(a): #len(a)=5
    print(a[i])
    printarr(a, i+1)

printarr([10,20,30,40,50], 0)
```

```python
# reverse array print
def printarr(a,i): #i=0,1,2,3,4
  if i<len(a): #len(a)=5
    printarr(a, i+1)
    print(a[i])

printarr([10,20,30,40,50], 0)
```

```python
# Sum of array
def sumarr(a, i):
  if i<len(a):
    return a[i] + sumarr(a, i+1)
  else:
    return 0

a= 10,20,30,40,50
print(sumarr(a, 0))
```

```python
#count / len
def sumarr(a, i):
  if i<len(a):
    return 1 + sumarr(a, i+1)
  else:
    return 0

a= 10,20,30,40,50
print(sumarr(a, 0))
```

```python
#factorial
def factorial(n):
  if n==0:
    return 1
  return n*factorial(n-1)

print(factorial(5))
```

```python
#summation
def summation(n):
  if n==0:
    return 0
  return n+summation(n-1)

print(summation(4))
```

```python
# decimal to binary
def binary(n):
  if n>0:
    binary(n//2)
    print(n%2, end = '')



binary(4)
```

```python
#fibonacci
def fib(n):
  if n <2:
    return n
  return fib(n-1) + fib(n-2)

fib(5)
```
