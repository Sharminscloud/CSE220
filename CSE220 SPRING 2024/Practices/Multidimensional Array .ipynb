```python
#sum all
import numpy as np
m = np.zeros((4,4), dtype = int)
arr=np.array([[0,2,2,0],
              [50,1,2,0],
              [2,2,2,0],
              [1,100,2,0]])
sum = 0
row, col = arr.shape
for i in range(row):
  for j in range(col):
    m[i][j]=arr[i][j]
    sum+= m[i][j]
print(sum)

```

```python
#row sum
import numpy as np
m = np.zeros((4,4), dtype = int)
arr=np.array([[0,2,2,0],
              [50,1,2,0],
              [2,2,2,0],
              [1,100,2,0]])
sumrow = 0
row, col = arr.shape
sum = np.zeros((row,1), dtype = int)
for i in range(row):
  for j in range(col):
    m[i][j]=arr[i][j]
    sumrow+= m[i][j]
    sum[i] =  sumrow
  sumrow = 0
print(sum)
```

```python
#column sum
import numpy as np
m=np.array([[0,2,2,0],
              [50,1,2,0],
              [2,2,2,0],
              [1,100,2,0]])
row, col = m.shape
m2 = np.zeros((1, col), dtype = int)
for i in range(col):
  for j in range(row):
    m2[0][i]+= m[j][i]
print(m2)
```
