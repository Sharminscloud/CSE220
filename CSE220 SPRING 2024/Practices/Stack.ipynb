```python
class Node:
  def __init__(self, element, next):
    self.element = element
    self.next = next
```

2 exceptions underflow and overflow
but overflow can only be in array as they have fixed len but we will use linked list here
and underflow can be in both array and linked list

```python
class Stack:
  def __init__(self):
    self.top = None  #self.top = Node(None, None)

  def push(self, val):
    new_node = Node(val, None)
    new_node.next = self.top
    self.top = new_node

  def pop(self):  #stack empty
    if self.top == None:
      return 'underflow'
    elem = self.top.element
    self.top = self.top.next
    return elem

  def peek(self):
    if self.top == None:
      return 'underflow'
    return self.top.element

  def is_empty(self):
    return (self.top==None)
```
