```python
class DoublyNode:
  def __init__(self, elem, next, prev):
    self.elem= elem
    self.next= next
    self.prev= prev

```

```python
a = [10, 20, 30, 40, 50, 60]
def createlistwithdoubly(a):
  dh = DoublyNode(None, None, None)
  dh.next = dh
  dh.prev = dh
  tail = dh

  for i in range(len(a)):
    n = DoublyNode(a[i], dh, tail)
    tail.next = n
    tail = tail.next
    dh.prev = tail
  return  dh
```

```python
def iteration(dh):
  temp = dh.next
  while temp!= dh:
    print(temp.elem, end=' ')
    temp= temp.next
  # return head

dh = createlistwithdoubly(a)
iteration(dh)
```
