```python
import numpy as np
def resizeArray(arr1, l):
  # arr2= np.zeros(l)
  arr2= np.array([0]* l)
  for i in range(len(arr1)):
    arr2[i]= arr1[i]
  return arr2


arr1= [1, 2, 3, 4]
print(resizeArray(arr1, 6))
# x= resizeArray(arr1, 6)
```

```python
#inplace
import numpy as np
a = np.array([10, 20, 30, 40]) #[40, 30, 20, 10]
count=0
half = len(a)//2
for i in range(len(a)-1, -1, -1):
  temp = a[i]
  a[i]=a[count]
  a[count]=temp
  count+=1
  if count==half:
    break
print(a)
```

```python
#selection sort
import numpy as np
a = np.array([9, 3, 6, 7])
for j in range(len(a)-1):
  for i in range(len(a)-1):
    temp = a[i]
    if a[i]<a[i+1]:
      a[i] = a[i+1]
      a[i+1] = temp
print(a)
```
