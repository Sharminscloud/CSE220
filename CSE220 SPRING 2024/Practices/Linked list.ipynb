```python
class Node:
  def __init__(self, e, n):
    self.elem=e
    self.next=n
```

```python
a = [10, 20, 30, 40, 50, 60]
def createlist(a):
  head = Node(a[0], None)
  tail = head
  for i in range(1, len(a)):
    n = Node(a[i], None)
    tail.next = n
    tail= tail.next
  return head
```

```python
def iteration(head):
  temp = head
  while temp!= None:
    print(temp.elem, end=' ')
    temp= temp.next
  # return head


head = createlist(a)
iteration(head)
```

```python
def countelem(head): #count is just len
  count = 0
  temp = head
  while temp!= None:
    count+=1
    temp = temp.next
  return count

print(countelem(head))
```

```python
def elemAt(head, idx):
  count = -1
  temp = head
  while temp != None:
    count+=1
    if count == idx:
      return (temp.elem)
    temp=temp.next
  # return head

print(elemAt(head, 3))
```

```python
def nodeAt(head, idx):
  count = -1
  temp = head
  while temp != None:
    count+=1
    if count == idx:
      return temp
    temp=temp.next
  # return head

print(nodeAt(head, 3))
print(nodeAt(head, 3).elem)
```

```python
def indexof (head, elem=int):
  temp = head
  index =-1
  while temp != None:
    index+=1
    if temp.elem == elem:
      return index
    temp=temp.next
  # return -1

print(indexof(head, 50))
```

```python
def removelem(head, idx):
  count = 0
  temp = head
  while temp != None:
    count+=1
    if count == idx:
      temp.next= temp.next.next
    print(temp.elem, end = ' ')
    temp=temp.next
  # return head

print(removelem(head, 3))
```

```python
def addlem(head, idx, elem):
  nnode = Node(elem, None)
  count = 0
  temp = head
  while temp != None:
    count+=1
    if count == idx:
      nnode.next = temp.next
      temp.next = nnode
    print(temp.elem, end = ' ')
    temp=temp.next
  # return head

print(addlem(head, 2, 0))
```
