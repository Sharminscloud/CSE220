```python
#Assignment Part 1
class Patient:
  #write a constructor
  def __init__(self, id, name, age, bloodgroup):
    self.id = id
    self.name = name
    self.age = age
    self.bloodgroup = bloodgroup
```

```python
class DoublyNode:
  def __init__(self, elem, next=None, prev=None):
    self.elem = elem
    self.next = next
    self.prev = prev

class WRM:
  def __init__(self):
    self.head = DoublyNode(None)
    self.size = 0

  def registerPatient(self, id, name, age, bloodgroup):
    new_patient = Patient(id, name, age, bloodgroup)
    new_node = DoublyNode(new_patient)
    if self.size == 0:
      self.head.next = new_node
      self.head.prev = new_node
      new_node.prev = self.head
      new_node.next = self.head
    else:
      last_node = self.head.prev
      new_node.next = self.head
      new_node.prev = last_node
      last_node.next = new_node
      self.head.prev = new_node
    self.size += 1 # Add the size of the waiting room
    print("Patient registered successfully.")

  def servePatient(self):
    if self.size == 0:
      print("No patients in the waiting room.")
      return
    first_patient = self.head.next.elem
    self.head.next = self.head.next.next
    self.head.next.prev = self.head
    self.size -= 1
    print("Now serving - ", "ID:", first_patient.id, "Name:", first_patient.name)

  def showAllPatient(self):
    if self.size == 0:
      print("No patients in the waiting room.")
      return
    current_node = self.head.next #as we used doubly
    print("Waiting patients:")
    while current_node != self.head:
      print("ID:", current_node.elem.id, "Name:", current_node.elem.name, "Age:", current_node.elem.age, "Blood Group:", current_node.elem.bloodgroup)
      current_node = current_node.next

  def canDoctorGoHome(self):
    if self.size == 0:
      print("Yes, the doctor can go home.")
    else:
      print("No, there are patients waiting.")

  def cancelAll(self):
    if self.size == 0:
      print("No appointments to cancel.")
      return
    self.head.next = self.head
    self.head.prev = self.head
    self.size = 0
    print("All appointments canceled. Doctor can go to lunch.")

  def ReverseTheLine(self):
    if self.size == 0:
      print("No patients in the waiting room.")
      return
    current_node = self.head.prev
    print("Reversed waiting patients names:")
    while current_node != self.head:
        print(print("ID:", current_node.elem.id, "Name:", current_node.elem.name, "Age:", current_node.elem.age, "Blood Group:", current_node.elem.bloodgroup))
        current_node = current_node.prev


  # def isEmpty(self):
  #   if self.dh==self.dh.next:
  #     return True
  #   else:
  #     return False
```

**Assignment Part 2: Stack**

Linked List based Stack is implemented in the following cell.

```python
class Node:
  def __init__(self,elem=None,next=None):
    self.elem = elem
    self.next = next

class Stack:
  def __init__(self):
    self.__top = None

  def push(self,elem):
    nn = Node(elem,self.__top)
    self.__top = nn

  def pop(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    e = self.__top
    self.__top = self.__top.next
    return e.elem

  def peek(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    return self.__top.elem

  def isEmpty(self):
    return self.__top == None
```

Task 1: Diamond Count

```python
def diamond_count(stack,string):
  #TO DO
  count=0
  for i in string:
    if i == '<':
      stack.push(i)
    if stack.peek() == '<' and i == '>':
      stack.pop()
      count+=1
  return count


print('Test 01')
stack = Stack()
string = '<..><.<..>> '
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 3
unittest.output_test(returned_value, 3)
print('-----------------------------------------')


print('Test 02')
stack = Stack()
string = '<<<..<......<<<<....>'
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 1
unittest.output_test(returned_value, 1)
print('-----------------------------------------')


print('Test 03')
stack = Stack()
string = '>>><...<<..>>...>...>>>'
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 3
unittest.output_test(returned_value, 3)
print('-----------------------------------------')
```

```python
def conditional_reverse(stack):
  #To Do
  newstack = Stack()
  while stack.peek() is not None:
    a = stack.pop()
    # print(a)
    if a == stack.peek():
      newstack.push(a)
      stack.pop()
    else:
      newstack.push(a)
  return newstack

print('Test 01')
st=Stack()
st.push(10)
st.push(10)
st.push(20)
st.push(20)
st.push(30)
st.push(10)
st.push(50)
print('Stack:')
print_stack(st)
print('------')
reversed_stack=conditional_reverse(st)
print('Conditional Reversed Stack:')
print_stack(reversed_stack) # This stack contains 50, 10, 30, 20, 10 in this order whereas top element should be 10
print('------')
```
