```python
class patient:
  def __init__(self, next, prev, id, name, age, bloodgroup):
    self.next = next
    self.prev = prev
    self.id = id
    self.name = name
    self.age = age
    self.bloodgroup = bloodgroup


class WRM:
  def __init__(self):
    self.head = patient(None, None, None, None, None, None) #for 6 init parameter var
    self.tail = self.head
    self.head.next = self.head #to make circuler
    self.head.prev = self.head

  def RegisterPatient(self, id, name, age, bloodgroup):
    new = patient( None, None, id, name, age, bloodgroup) #none for 2 not included
    dum = self.head.prev
    self.head.prev = new
    new.next = self.head
    dum.next = new
    new.prev = dum
    print( 'Patient name: ', name, ", your registration is successful!")

  def ServePatient(self):
    dhead = self.head
    if self.head.next != self.head:
      print('patient', self.head.next.name, 'is served!')
      self.head.next = self.head.next.next
      self.head.next.next.prev = self.head
    else:
        print("No patients to serve.")

  def CancelAll(self):
    self.head.next = self.head #circuler
    self.head.prev = self.head
    print("Cancelling all appoinments, have a good lunch!.")

  def CanDoctorGoHome(self):
      if self.head.next != self.head:
          return False
      else:
          return True

  def ShowAllPatient(self):
    current = self.tail
    num = 1
    if current.next == self.head:
      print("No patients are in the waiting room.")
    else:
      while current.next != self.head:
          print(num, 'patient is in the waiting room')
          current = current.next
          num+=1

#drivercode
manage = WRM()
print("Welcome to Waiting Room Management!")
while True:
  print('\n---------------------------------------')
  print("Hello there! Please choose your option, Thank you.")
  print('\n 1. Add Patient\n 2. Serve Patient\n 3. Cancel all Appointment\n 4. Can Doctor go Home?\n 5. Show All patients.\n 6. Exit')
  enter = int(input("Your choosen option no: "))
  if enter ==1:
    id = int(input('Please enter patients id: '))
    name = input('Please enter patients name: ')
    age = int(input('Please enter patients age: '))
    blood = input('Please enter patients blood group: ')
    manage.RegisterPatient(id, name, age, blood)
  elif enter ==2:
    manage.ServePatient()
  elif enter == 3:
    manage.CancelAll()
  elif enter == 4:
    if manage.CanDoctorGoHome() == True:
        print('Yes!')
    else:
        print('NO, Sorry.')
  elif enter == 5:
    manage.ShowAllPatient()
  elif enter == 6:
    print("Thank you for using our management system!")
    break
print('---------------------------------------')
```
