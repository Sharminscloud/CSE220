```python
#You must run this cell to print matrix and for the driver code to work
def print_matrix(m):
  row,col = m.shape
  for i in range(row):
    c = 1
    print('|', end='')
    for j in range(col):
      c += 1
      if(len(str(m[i][j])) == 1):
        print(' ',m[i][j], end = '  |')
        c += 6
      else:
        print(' ',m[i][j], end = ' |')
        c += 6
    print()
    print('-'*(c-col))

```

Task 1: Zigzag Walk

```python
def walk_zigzag(floor):
    row, column = floor.shape #for matrix row column number
    for tiles in range(row):
      if tiles % 2 != 0:
        for evenstep in range(column-1, -1, -2): #5, -1, -2
          print(floor[tiles][evenstep],end=' ')
      if tiles % 2 == 0:
        for oddstep in range(0, column, 2): #0, 6, 2
          print(floor[tiles][oddstep],end=' ')
      print() #to match the displayed output


floor = np.array([[ '3W' , '8B' , '4W' , '6B' , '1W' , '5B'],
                  ['3B' , '2W' , '1B' , '6W' , '3B' , '8W'],
                  ['9W' , '0B' , '7W' , '5B' , '3W' , '8B'],
                  ['2B' , '1W' , '3B' , '6W' , '0B' , '4W'],
                  ['1W' , '4B' , '2W' , '8B' , '6W' , '6B']]
                )
print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3W 4W 1W
# 8W 6W 2W
# 9W 7W 3W
# 4W 6W 1W
# 1W 2W 6W
```

Task 2: Landscape Screen

```python
def landscape(m):
  row, column = m.shape #4,3
  newarray= np.zeros((column, row), dtype=int) #creates array of column and row length
  for num1 in range(row):
    for num2 in range(column):
      newarray[num2][num1] = m[num1][num2] #changes col with row and row with col
  return newarray


m = np.random.randint(15,size=(4,3)) #This creates an np array of 4 x 3 where the values are random numbers from [0,15) range
print_matrix(m)
print()
print_matrix(landscape(m))
```

Task 3: Seating Arrangement

```python
import numpy as np
def arrangement(m, diff):
  row, column= m.shape #4, 4
  for num1 in range(len(m)-1): #4-1=3
    for num2 in range(len(m[num1])): #3
        if num1 != num2:
          substraction= m[num1][num2] - m[num2][num1] #r1col2 and col1row2 substraction
          if substraction < 0: #cant be negative
            substraction =- substraction
            if substraction != diff: #have to match
              return False
  return True

m = np.array([[7,13,9,14],[12,8,15,11],[10,17,3,2],[15,10,1,4]])
print_matrix(m)
print()
returned_value = arrangement(m,1)
print(returned_value)#This should print False
unittest.output_test(returned_value, False)

m = np.array([[8,15,7,12],[13,2,18,11],[9,20,5,2],[14,9,0,10]])
print_matrix(m)
print()
returned_value = arrangement(m,2)
print(returned_value)#This should print True
unittest.output_test(returned_value, True)
```

Task 4: Chess Piece

```python
def show_knight_move(knight):
  row, column= knight[0],knight[1] #so row=3 and col=4
  newarray = np.zeros((8, 8), dtype=int)
  newarray[row,column]= 33 #position of the knight
  #updown
  if column+2<8: #<8 so that it stays in the shape
    if row+1<8:
      newarray[row+1,column+2]=3
    if row-1<8:
      newarray[row-1,column+2]=3
  if column-2<8:
    if row+1<8:
      newarray[row+1,column-2]=3
    if row-1<8:
      newarray[row-1,column-2]=3
  #rightleft
  if row+2<8:
    if column+1<8:
      newarray[row+2,column+1]=3
    if column-1<8:
      newarray[row+2,column-1]=3
  if row-2<8:
    if column+1<8:
      newarray[row-2,column+1]=3
    if column-1<8:
      newarray[row-2,column-1]=3
  return newarray


knight = (3,4)
chess_board = show_knight_move(knight)
print_matrix(chess_board)
```
